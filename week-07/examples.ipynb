```python
import os
from dotenv import load_dotenv
from supabase import create_client
import pandas as pd
import plotly.express as px

load_dotenv()

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_ANON_KEY")

if not SUPABASE_URL or not SUPABASE_KEY:
    raise ValueError("Puudub SUPABASE_URL või SUPABASE_ANON_KEY .env failis")

supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

response = supabase.table("sales").select("*").execute()
df = pd.DataFrame(response.data)

print(df.head())
print(df.info())

df["sale_date"] = pd.to_datetime(df["sale_date"])

daily_sales = df.groupby("sale_date")["total_price"].sum().reset_index()

fig = px.line(
    daily_sales,
    x="sale_date",
    y="total_price",
    title="Müük ajas",
)

fig.show()
```

```python
response = supabase.table("sales").select("*").execute()
sales_df = pd.DataFrame(response.data)

response = supabase.table("products").select("*").execute()
products_df = pd.DataFrame(response.data)

merged_df = sales_df.merge(
    products_df,
    on="product_id",
    suffixes=("_sale", "_product")
)


print(merged_df.head())
```
